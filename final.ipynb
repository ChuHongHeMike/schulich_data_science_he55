## Initial hypothesis

Based on the EDA, we can see the Age, Location, Device, and the purchased items by time will be the 4 factors to predict the customer who will churn. This is because they have the relative high correlation on 'churn'.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc, precision_recall_curve
x = df.drop(['Churn','Unnamed: 0','CustomerID'],axis=1)
y = df['Churn']
rf = RandomForestClassifier(n_estimators = 3, max_depth=3)


rf.fit(x,y)
```

```python
importances = rf.feature_importances_
feature_names = x.columns

# Sort feature importances in descending order and get their indices
indices = np.argsort(importances)[::-1]


plt.figure(figsize=(10, 5))
plt.title("Feature Importance")
plt.bar(feature_names[indices], importances[indices])
plt.xticks(feature_names[indices], rotation=90)

plt.show()
print(feature_names[indices])
print(importances[indices])
```

```python
from scipy import stats

churn_Yes= df[df['Churn'] == 1]['Location']
churn_No= df[df['Churn'] == 0]['Location']

t_stat, p_val = stats.ttest_ind(churn_Yes, churn_No)

print(f"F-statistict Location: {t_stat}")
print(f"P-value Location: {p_val}")
print('-------------------------------------------')
print(df.groupby('Churn')[['Location']].mean())
print('-------------------------------------------')

churn_Yes= df[df['Churn'] == 1]['Device']
churn_No= df[df['Churn'] == 0]['Device']

t_stat, p_val = stats.ttest_ind(churn_Yes, churn_No)

print(f"F-statistict Device: {t_stat}")
print(f"P-value Device: {p_val}")
print('-------------------------------------------')
print(df.groupby('Churn')[['Device']].mean())
print('-------------------------------------------')
churn_Yes= df[df['Churn'] == 1]['Age']
churn_No= df[df['Churn'] == 0]['Age']

t_stat, p_val = stats.ttest_ind(churn_Yes, churn_No)

print(f"F-statistict Device: {t_stat}")
print(f"P-value Device: {p_val}")
print('-------------------------------------------')
print(df.groupby('Churn')[['Age']].mean())
print('-------------------------------------------')
```

Since both Location and Device p-values are lower than 0.05, the company can conclude that the Location and Device of those who churned are statistically different from the customer who are living in different areas and not churned. This conclusion may also help the business to identify potential high-churn or low-churn customers accordingly to manage risk and improve the overall business performance based on their differences.

And we can see that 'Location', 'Device', and 'Age' have the top three correlation with 'Churn'. The following is the number of items purchased in each month. Also, there are not a significant difference between spend, purchases, and the number of calling customer services.
