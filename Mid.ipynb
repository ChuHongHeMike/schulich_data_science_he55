```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('C:/data_science/midterm_data.csv') #loading dataset
df
```

```python
df.info()
df.describe()
```

```python
df[df['device'].isnull() & df['location'].isnull()]
```

```python
df1 = df
df1.drop(df1[df1['device'].isnull() & df1['location'].isnull()].index, axis=0, inplace=True)
```

```python
df1.info()
df1.describe()
```

```python
df2 = df.copy()
device_list = df2['device'].unique()
device_list = device_list[1:]
df2['device'].isnull().sum()
```

```python
random_device = np.random.choice(device_list, size = df2['device'].isnull().sum())
df1.loc[df1['device'].isnull(), 'device'] = random_device
```

```python
df1.info()
```

```python
df2 = df.copy()
location_list = df2['location'].unique()
location_list = location_list[1:]
df2['location'].isnull().sum()
```

```python
random_location = np.random.choice(location_list, size = df2['location'].isnull().sum())
df1.loc[df1['location'].isnull(), 'location'] = random_location
```

```python
df1.info()
```

```python
var1 = df1[df1['checkout_status']==1]['cart_items']
var2 = df1[df1['checkout_status']==0]['cart_items']
p_val = stats.ttest_ind(var1, var2)
print(p_val)
```

```python
df1.groupby('checkout_status')[['cart_items']].mean()
```

```python
var1 = df1[df1['checkout_status']==1]['cart_value']
var2 = df1[df1['checkout_status']==0]['cart_value']
p_val = stats.ttest_ind(var1, var2)
print(p_val)
```

```python
df1.groupby('checkout_status')[['cart_value']].mean()
```

```python
df1["age_group"] = pd.cut(x=df1['age'], bins=[0,30,60,100], labels=["young","middle_aged","old"])
df1
```

```python
var1 = df1[df1['checkout_status']==1]['time_spent']
var2 = df1[df1['checkout_status']==0]['time_spent']
p_val = stats.ttest_ind(var1, var2)
print(p_val)
```

```python
df1.groupby('checkout_status')[['time_spent']].mean()
```
