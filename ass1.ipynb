```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
# get the htlm
resp = requests.get('https://ca.trustpilot.com/review/equitablebank.ca')

print(resp.text)
# This is the html source
html_code = resp.text
```

```python
soup = BeautifulSoup(html_code, 'html.parser')
item = soup.find(
name='p',
attrs={'class':'typography_body-l__KUYFJ typography_appearance-default__AAY17'}
)
N = int(item.contents[0].replace(',',''))
print(N)
```

```python

current = 'https://ca.trustpilot.com/review/equitablebank.ca'
page = 'https://ca.trustpilot.com'
All_page=[current]
while current:
    response = requests.get(current)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
    
        next_page = soup.find_all(name="a", string='Next page')

        if next_page[0].get('href', ''):
            next_url = page + next_page[0]['href']
            current = next_url
            print(current)
            All_page.append(current)
        else:
            current = None
    else:
        current = None
All_page
```

```python
data = []
for url in All_page:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    review = soup.find_all(
        name='p',
        attrs={
              'class':'typography_body-l__KUYFJ typography_appearance-default__AAY17 typography_color-black__5LYEn',
              'data-service-review-text-typography':'true'
               },
    )
    for paragraph in review:
            data.append(paragraph.text)
df = pd.DataFrame({'Text': data})
```
